# Inference + API Smoke Test

This notebook validates:
- Loading exported artifacts
- Running an inference call
- Exercising the FastAPI endpoint via an in-memory test client

```python
from healthcare_digital_twin.schemas import PredictionRequest
from healthcare_digital_twin.inference import predict

sample = PredictionRequest(
    Age=45,
    Gender=1,
    BMI=28.2,
    Systolic_BP=130,
    Diastolic_BP=82,
    Glucose=110,
    Insulin=12,
    Total_Cholesterol=190,
    HDL_Cholesterol=48,
)
predict(sample)
```

```python
from fastapi.testclient import TestClient
from healthcare_digital_twin.api import app

client = TestClient(app)
resp = client.post('/predict', json=sample.model_dump())
resp.status_code, resp.json()
```
