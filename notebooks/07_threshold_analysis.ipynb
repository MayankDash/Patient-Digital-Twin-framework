# Threshold Optimization Analysis

Reproduces the threshold sweep idea from your tuning notebook, but using the project module.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from healthcare_digital_twin.data import load_processed_patient_state
from healthcare_digital_twin.modeling import make_logreg_pipeline, threshold_sweep
from healthcare_digital_twin.constants import FEATURES, TARGET

df = load_processed_patient_state()
X = df[FEATURES]
y = df[TARGET].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
pipe = make_logreg_pipeline()
pipe.fit(X_train, y_train)
y_prob = pipe.predict_proba(X_test)[:, 1]
sweep = threshold_sweep(y_test.to_numpy(), y_prob)
sweep.head()
```

```python
plt.figure(figsize=(8,5))
plt.plot(sweep['threshold'], sweep['precision'], label='Precision')
plt.plot(sweep['threshold'], sweep['recall'], label='Recall')
plt.plot(sweep['threshold'], sweep['f1'], label='F1 Score')
plt.xlabel('Decision Threshold')
plt.ylabel('Score')
plt.title('Threshold Optimization Analysis')
plt.legend()
plt.tight_layout()
plt.show()
```
