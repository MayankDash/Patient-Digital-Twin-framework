# Train + Export (Logistic Regression)

This reproduces the paper pipeline using the project module:
- Stratified train/test split
- Median imputation + StandardScaler
- Logistic Regression (optional GridSearchCV tuned on F1)
- Export artifacts to `artifacts/` and plots to `reports/figures/`

```python
from healthcare_digital_twin.data import load_processed_patient_state
from healthcare_digital_twin.modeling import train_logistic_regression, save_artifacts, save_confusion_matrix_plot

df = load_processed_patient_state()
result, metadata = train_logistic_regression(df, tune=True, threshold=0.4)
result.metrics
```

```python
model_path, meta_path = save_artifacts(estimator=result.estimator, metadata=metadata, artifact_name='logreg_pipeline')
fig_path = save_confusion_matrix_plot(result.confusion_matrix, title='Confusion Matrix - Logistic Regression', out_name='logreg_pipeline.confusion_matrix.png')
model_path, meta_path, fig_path
```
